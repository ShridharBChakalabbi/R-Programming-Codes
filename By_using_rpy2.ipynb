```python
import rpy2
import rpy2.robjects as robjects
import numpy as np
```

```python
m = robjects.r('''
library(Lock5Data)
data(CommuteAtlanta)
str(CommuteAtlanta)
time.mean = with(CommuteAtlanta, mean(Time))
time.mean
B = 1000
n = nrow(CommuteAtlanta)
boot.samples = matrix(sample(CommuteAtlanta$Time, size = B * B, replace = TRUE),B, B)
        ''')
```

```python
res = robjects.r('''
results<-vector()
for (i in 1:1000)
{
  results[i] = mean(boot.samples[,i])
}
#print(mean(results))
#print(sd(results))
result = results
''')
```
